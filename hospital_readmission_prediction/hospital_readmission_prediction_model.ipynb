# Preprocessing

```python
data.isnull().sum()
```

```python
((data['A1C_Result'].isnull().sum())/data.shape[0])*100
```

```python
data1=data.copy()
```

```python
data1['A1C_Result'].fillna('Unknown', inplace=True)
```

```python
data1.isnull().sum()
```

```python
data1.info()
```

```python
print(data1['Gender'].unique())
print(data1['Admission_Type'].unique())
print(data1['Diagnosis'].unique())
print(data1['A1C_Result'].unique())
print(data1['Readmitted'].unique())
```

```python
col =['Gender', 'Admission_Type', 'Diagnosis','A1C_Result']

from sklearn.preprocessing import LabelEncoder

for i in col:
    
    label_encoder = LabelEncoder()
    data1[i] = label_encoder.fit_transform(data1[i])
```

```python
data1.info()
```

```python
a = {'No':0,'Yes':1}

data1['Readmitted'] = data1['Readmitted'].map(a)
```

```python
data1.head()
```

```python
#removing outliers
q1=data1['Age'].quantile(0.25)
q3=data1['Age'].quantile(0.75)

IQR=q3-q1

lower_bound = q1-1.5*IQR
upper_bound = q3+1.5*IQR

data1 = data1[(data1['Age'] >= lower_bound) & (data1['Age'] <= upper_bound)]
```

```python
data1.shape
```

# Feature Engineering

```python
bins = [0, 18, 35, 60, 80, 120]
labels = ['Child', 'Young Adult', 'Adult', 'Senior', 'Elderly']
data1['Age_Group'] = pd.cut(data1['Age'], bins=bins, labels=labels, right=False)
```

```python
data1['Age'].unique()
```

```python
data1['Age_Group'].unique()
```

```python
b={'Child':0,'Young Adult':1,'Adult':2,'Senior':3,'Elderly':4}
data1['Age_Group']=data1['Age_Group'].map(b)
```

```python
data1['Age_Group'].unique()
```

```python
data1['Total_Visits'] = (
    data1['Num_Outpatient_Visits'] +
    data1['Num_Inpatient_Visits'] +
    data1['Num_Emergency_Visits'])

```

```python
data1['High_Diagnosis_Count'] = (data1['Num_Diagnoses'] > 5).astype(int)
```

```python
data1.head()
```

```python
data1 = data1.drop(columns=['Patient_ID','Num_Outpatient_Visits','Num_Inpatient_Visits','Num_Emergency_Visits','Num_Diagnoses'])
```

# Model selection & training

```python
# Define features and target
x = data1.drop(columns=['Readmitted'])  
y = data1['Readmitted']
```

```python
#  Train-Test Split
from sklearn.model_selection import train_test_split, GridSearchCV

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier( n_neighbors=7)
knn.fit(x_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay,accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, roc_curve, classification_report

y_pred_knn =knn.predict(x_test)
y_pred_proba_knn =knn.predict_proba(x_test)[:,1]

# Accuracy
accuracy_knn = accuracy_score(y_test, y_pred_knn)

# ROC-AUC
roc_auc_knn = roc_auc_score(y_test, y_pred_proba_knn)

print(f"Accuracy: {accuracy_knn:.2f}")
print(f"ROC-AUC: {roc_auc_knn:.2f}")
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(max_iter=10000)
lr.fit(x_train,y_train)
```

```python
y_pred_lr =lr.predict(x_test)
y_pred_proba_lr =lr.predict_proba(x_test)[:,1]

# Accuracy
accuracy_lr = accuracy_score(y_test, y_pred_lr)

# ROC-AUC
roc_auc_lr = roc_auc_score(y_test, y_pred_proba_lr)

print(f"Accuracy: {accuracy_lr:.2f}")
print(f"ROC-AUC: {roc_auc_lr:.2f}")
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier()
rfc.fit(x_train,y_train)
```

```python
y_pred_rfc =rfc.predict(x_test)
y_pred_proba_rfc =rfc.predict_proba(x_test)[:,1]

# Accuracy
accuracy_rfc = accuracy_score(y_test, y_pred_rfc)

# ROC-AUC
roc_auc_rfc = roc_auc_score(y_test, y_pred_proba_rfc)

print(f"Accuracy: {accuracy_rfc:.2f}")
print(f"ROC-AUC: {roc_auc_rfc:.2f}")
```

```python
# as we can see that we got higher accuracy with the logistic regression, so we will select it as our model.
```
